```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
atual = pd.read_excel('output.xlsx', sheet_name='Base')
```

```python
atual.replace(np.nan, 0, inplace=True)
```

```python
atual
```

```python
atual_narratives = atual.loc[:, 'submissão-recato-passividade':'objetificação, mulher objeto'].sum()
```

```python
atual_narratives.sum()
```

```python
atual_narratives
```

```python
atual_narratives_perc = atual_narratives.apply(lambda x: x/atual_narratives.sum())
```

```python
atual_narratives_perc
```

```python
mpb = pd.read_csv('mpb.csv')
```

```python
mpb = mpb.set_index('Unnamed: 0')
```

```python
mpb_narratives = mpb['0']
```

```python
mpb_narratives.sum()
```

```python
mpb_narratives
```

```python
mpb_narratives_perc = mpb_narratives.apply(lambda x: x/mpb_narratives.sum())
```

```python
mpb_narratives_perc
```

```python
df = pd.concat([mpb_narratives_perc, atual_narratives_perc], axis=1)
```

```python
df.columns = ['Anos 70', 'Setembro 2022']
```

```python
df.replace(np.nan, 0, inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={'index':'Narrativa'}, inplace=True)
```

```python
df_melt = df.melt('Narrativa', var_name='Período', value_name='Frequência em porcentagem de aparições (%)')
```

```python
sns.set_style('whitegrid')
```

```python
fig = plt.figure()
fig.set_size_inches(6, 10)
fig.patch.set_facecolor('w')
fig.suptitle('Evolução da frequência de narrativas sexistas (anos 70 - 2022)')
fig.set_dpi(300)
sns.barplot(x='Frequência em porcentagem de aparições (%)', y='Narrativa', hue='Período', data=df_melt, palette='RdGy')
plt.show()
```

```python
df
```

```python
dif = df.loc[:, 'Anos 70':'Setembro 2022'].pct_change(axis='columns')
```

```python
dif.rename(columns={'Setembro 2022':'Diferença'}, inplace=True)
```

```python
df['Alteração'] = dif['Diferença']
```

```python
df.sort_values('Alteração', inplace=True)
```

```python
df
```

```python
df_plt = df.loc[df.index<14]
```

```python
df_plt
```

```python
fig = plt.figure()
sns.barplot(x=df_plt['Alteração'], y=df_plt['Narrativa'], palette='Reds')
fig.patch.set_facecolor('w')
fig.suptitle('Mudança percentual na frequência das narrativas entre anos 70 e setembro de 2022')
fig.set_dpi(300)
plt.show()
```
