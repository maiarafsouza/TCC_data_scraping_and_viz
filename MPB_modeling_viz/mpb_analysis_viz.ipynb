```python
n_list = ['submissão_recato_passividade', 'racionalidade_x_afetividade', 'vingança', 'abandono', 'maior_consciência_autônomia', 'culpabilidade_pelo_sofrimento_do_homem', 'discurso_da_espera_feminina', 'idealização_idolatração_mulher_musa', 'misoginia', 'homem_aconselhador', 'oportunismo_falsidade', 'mulher_doméstica', 'fragilidade_feminina', 'santa_x_prostituta']
```

```python
n_dict = {'submissão-recato-passividade':submissão_recato_passividade, 'racionalidade x afetividade':racionalidade_x_afetividade, 'vingança':vingança, 'abandono':abandono, 'maior consciência, autônomia (ruptura com o discurso dominante)':maior_consciência_autônomia, 'culpabilidade pelo sofrimento do homem':culpabilidade_pelo_sofrimento_do_homem, 'imanência x transcedência, discurso da espera feminina':discurso_da_espera_feminina, 'idealização, idolatração, mulher-musa':idealização_idolatração_mulher_musa, 'misoginia, feminino menosprezado, incapacidade intelectual':misoginia, 'discurso da verdade proferido pelo homem, aconselhador':homem_aconselhador, 'oportunismo e falsidade':oportunismo_falsidade, 'mulher relacionada a esfera doméstica':mulher_doméstica, 'fragilidade feminina':fragilidade_feminina, 'dicotomia santa x prostituta':santa_x_prostituta}
```

```python
for key, value in n_dict.items():
    df[key] = df['song_title'].isin(value)
```

```python
df.replace({False:0, True:1}, inplace=True)
```

```python
df
```

```python
narrative_sum = df.sum(numeric_only=True)
```

```python
narrative_sum.sort_values(ascending=False)
```

```python
narrative_sum_plt = narrative_sum.sort_values()
```

```python
narrative_sum_plt = pd.DataFrame(narrative_sum_plt)
```

```python
narrative_sum_plt.reset_index(inplace=True)
```

```python
narrative_sum_plt.rename(columns={'index':'narrativa', 0:'frequência'}, inplace=True)
```

```python
sns.set_style('whitegrid')
sns.set_palette(palette='Reds')
```

```python
fig = plt.figure()
sns.barplot(y='narrativa', x='frequência', data=narrative_sum_plt, palette='Reds')
fig.patch.set_facecolor('w')
fig.set_size_inches(6, 10)
fig.suptitle('Frequência das narrativas sexistas (MPB, anos 70)')
fig.set_dpi(300)
plt.show()
```

```python
song_sum = df.set_index('song_title').sum(axis=1)
```

```python
song_sum_plt = song_sum.sort_values()
song_sum_plt = pd.DataFrame(song_sum_plt)
song_sum_plt.reset_index(inplace=True)
```

```python
p = [sns.color_palette(palette='Reds')[5], sns.color_palette(palette='Reds')[5]]
fig = plt.figure()
sns.barplot(y='música', x='quantidade de narrativas', data=song_sum_plt, palette=p)
fig.patch.set_facecolor('w')
fig.set_size_inches(4, 20)
fig.suptitle('Quantidade de narrativas sexistas por música (MPB, anos 70)')
fig.set_dpi(300)
plt.show()
```
