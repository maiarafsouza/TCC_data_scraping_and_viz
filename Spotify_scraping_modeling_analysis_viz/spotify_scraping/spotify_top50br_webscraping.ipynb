```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime
import pandas as pd
```

```python
top50_file = requests.get('https://open.spotify.com/playlist/37i9dQZEVXbMXbN3EUUhlg').text
top50_soup = BeautifulSoup(top50_file, 'lxml')
print(top50_soup.prettify())
```

```python
today = datetime.today()
today = f'{today.day}_{today.month}_{today.year}'
print(today)
```

```python
song_rank = []
for song in top50_soup.find_all(class_="Type__StyledComponent-sc-1ell6iv-0 MestoBold-sc-16x8d2-0 EntityRowV2__RowNumber-sc-ayafop-7 hIUPce fYiTde"):
    song_rank.append(song.text)
print(song_rank)
```

```python
song_link = []
set = top50_soup.find_all(class_='EntityRowV2__AnchorLink-sc-ayafop-8 cGeiev')
for song in set:
    song_link.append(song.get('href'))
print(song_link)
```

```python
song_title = []
song_author = []

for song in song_link:
    song_file = requests.get(f'{song}').text
    song_soup = BeautifulSoup(song_file, 'lxml')
    song_title.append(song_soup.title.text.split('|')[0])
    song_author.append(song_soup.find(class_='Type__TypeElement-goli3j-0 iTnRIT T3C0eBxRoJcpqMNCpLT6 FyfbzzWIE70wBv67eCsB').text)

print(song_title)
print(song_author)
```

```python
idx = len(song_rank)
print(f'{idx} songs')
```

```python
rank = {'song_rank':song_rank, 'song_title':song_title, 'song_author':song_author, 'song_link':song_link}
df = pd.DataFrame(rank)
```

```python
df.set_index('song_rank')
```

```python
df.to_excel(f'topbr_spotify_{today}.xlsx', index='song_rank')
df.to_csv(f'topbr_spotify_{today}.csv', index='song_rank')
```
