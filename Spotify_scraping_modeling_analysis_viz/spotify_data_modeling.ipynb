```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
files = [file for file in os.listdir('spotify_scraped_data')]
```

```python
df = pd.DataFrame()
```

```python
for file in files:
    f = pd.read_csv(f'data/{file}')
    df = pd.concat([df, f])
```

```python
pd.set_option('display.max_rows', 100)
```

```python
df
```

```python
df['freq'] = np.nan
```

```python
for song_title in df['song_title'].unique():
    filt = df['song_title']==f'{song_title}'
    for i in df.loc[filt].index:
        df.loc[i, 'freq'] = df.loc[filt].shape[0]
```

```python
df
```

```python
df['freq'].value_counts()
```

```python
df['rank_sum'] = np.nan
```

```python
for s in df['song_title'].unique():
    filt = df['song_title']==f'{s}'
    for i in df.loc[filt].index:
        df.loc[i, 'rank_sum'] = df.loc[df['song_title']==f'{s}']['song_rank'].sum()
```

```python
df_new = pd.DataFrame({'song_title':df['song_title'], 'song_author':df['song_author'], 'song_link':df['song_link'], 'freq':df['freq'], 'rank_sum':df['rank_sum']})
```
