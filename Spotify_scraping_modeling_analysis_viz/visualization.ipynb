```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('output.xlsx', sheet_name='Base')
```

```python
# Soma por arquétipo, Soma por música, Citações
```

```python
df.replace(np.nan, 0, inplace=True)
```

```python
df
```

```python
sns.set_style('whitegrid')
sns.set_palette(palette='Reds')
```

```python
song_count = df.loc[df['total']>0].sort_values('total')
```

```python
p = [sns.color_palette(palette='Reds')[5], sns.color_palette(palette='Reds')[5]]
```

```python
fig = plt.figure()
sns.barplot(y='song_title', x='total', data=song_count, palette=p)
fig.patch.set_facecolor('w')
fig.set_size_inches(4, 20)
fig.suptitle('Quantidade de narrativas sexistas por música (atual)')
fig.set_dpi(300)
plt.show()
```

```python
ar = df.loc[:, 'submissão-recato-passividade':'objetificação, mulher objeto']
```

```python
ar = ar.transpose()
```

```python
ar
```

```python
ar['total'] = np.nan
```

```python
for i in ar.index:
    ar.loc[f'{i}', 'total'] = ar.loc[f'{i}', :].sum()
```

```python
ar.sort_values('total', inplace=True)
```

```python
ar.reset_index(inplace=True)
```

```python
ar.rename(columns={'index':'narrativa'}, inplace=True)
```

```python
ar = ar.loc[ar['total']>0]
```

```python
fig = plt.figure()
p = sns.color_palette('Reds').reverse()
sns.barplot(y='narrativa', x='total', data=ar, palette='Reds')
fig.patch.set_facecolor('w')
fig.set_size_inches(6, 10)
fig.suptitle('Frequência das narrativas sexistas (atual)')
fig.set_dpi(300)
plt.show()
```
